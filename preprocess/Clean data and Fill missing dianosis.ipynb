```python
import sys
sys.path.append("../")
import os
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import misc
import math 
```

## LOAD DATA

```python
columns = ["RID", "PTID", "EXAMDATE", "Month", "DX_bl", "DX", "AGE", "PTGENDER", "APOE4", "PTEDUCAT",
           "CDRSB", "MMSE", "ADAS13", "RAVLT_immediate", "RAVLT_learning", "RAVLT_perc_forgetting", "FAQ",
           ]

data = pd.read_csv("../dataset/ADNI/ADNIMERGE.csv", usecols=columns)
print("Number of Patient: ", len(data['RID'].unique()))
data['MRI'], data['PET'], data['DTI'] = 0, 0, 0
data = data.set_index('RID')
print(len(data["PTID"].unique()), len(data))
data
```

## Filter timepoint with bl, 1 years, 2 years,..

```python
data = data[data["Month"] != 3]
data = data[data["Month"] %12 !=6]
data = data[data["Month"] <= 120]
data
```

# Find and ignore list of non-downloaded ID

```python
path = "D:/0.PhD/Project/ADPrediction/src/dataset/ADNI/renamed"
list_collected = os.listdir(path)
list_not_collected = []
data_collected = pd.DataFrame(None)
for idx, ele in enumerate(data["PTID"].unique()):
    if ele not in list_collected:
        list_not_collected.append(ele)
    else:
        data_collected = pd.concat([data_collected, data[data["PTID"]==ele]])
        
print(len(list_not_collected))
data_collected
```

## Drop duplicate

```python
features = ["PTID", "Month"]

data_collected = data_collected.drop_duplicates(subset=features, keep="first")
data_collected
```

## Remove reversible and missing at baseline cases

```python
print("Number of missing diagnosis: ", data_collected["DX"].isnull().sum())
print("Number of having diagnosis: ", len(data_collected) - data_collected["DX"].isnull().sum())
```

```python
list_patients_id = np.unique(data_collected.index)
num_patient = len(list_patients_id)
print("Number of patients: ", num_patient)
```

```python
n_wrong_dx = 0
list_id_wrong_dx = []
for rid in list_patients_id:
    if type(data_collected.loc[rid]["PTID"]) is str: # Only baseline timestamp
#         if data_collected.loc[rid]["DX"] != data_collected.loc[rid]["DX"]: #check null
        list_id_wrong_dx.append(rid)
        continue
    patientInfo = data_collected.loc[rid].sort_values(by="Month",ascending=True)
    ######## Check if patients have visits after 2 years or more from first visit 
    if not np.any(patientInfo["Month"] >= 24):
        list_id_wrong_dx.append(rid)
        continue
    #### Check if there are any labels except baseline
    if np.all(np.array(patientInfo["DX"])[1:] != np.array(patientInfo["DX"])[1:]):
        list_id_wrong_dx.append(rid)
        continue
    ####################
    old_dx = 'CN'
    list_dx = list(patientInfo['DX'])
    flag = False
    for idx in range(len(list_dx)):
        if list_dx[idx] != list_dx[idx]: # check nan
            continue
        else:
            rec_dx = list_dx[idx]
            if old_dx == "CN" and (rec_dx == "MCI" or rec_dx == "Dementia"):
                old_dx = rec_dx
            elif old_dx == "MCI" and rec_dx == "Dementia":
                old_dx = rec_dx
            elif old_dx == "MCI" and rec_dx == "CN":
                flag=True
            elif old_dx == "Dementia" and (rec_dx == "CN" or rec_dx == "MCI"):
                flag=True
        ## Check wrong diagnosis  
        if flag:   
            list_id_wrong_dx.append(rid)
            n_wrong_dx += 1
#             print(n_wrong_dx, rid, list_dx)
            break 
    
## Drop wrong dx
data_collected_cleaned = data_collected.drop(list_id_wrong_dx)
print(len(data_collected_cleaned.index.unique()), n_wrong_dx)
data_collected_cleaned
```

## Fill missing daignosis 

```python
list_patients_id = np.unique(data_collected_cleaned.index)
print(len(list_patients_id))
list_id_wrong_dx = []
data_filled_dx = pd.DataFrame(None)
for rid in list_patients_id:
    if type(data_collected_cleaned.loc[rid]["PTID"]) is str: # Only baseline timestamp
#         data_filled_dx = pd.concat([data_filled_dx, data_collected_cleaned.loc[rid:rid]])
#         print(rid)
        continue
    
    patientInfo = data_collected_cleaned.loc[rid].sort_values(by="Month",ascending=True)
    old_dx = 'CN'
    list_idx_miss = []
    list_dx = list(patientInfo.loc[rid]['DX'])
    for idx in range(len(list_dx)):
        if list_dx[idx] != list_dx[idx]: # check nan
            if old_dx == "Dementia":
                patientInfo.loc[rid]['DX'].iloc[idx] = old_dx
            else:
                list_idx_miss.append(idx)
        else:
            rec_dx = list_dx[idx]
            if len(list_idx_miss) > 0 and rec_dx == old_dx:
                for j in list_idx_miss:
                    patientInfo.loc[rid]['DX'].iloc[[j]] = old_dx
            list_idx_miss = [] #reset
            old_dx = rec_dx
    
    data_filled_dx = pd.concat([data_filled_dx, patientInfo])
    
    
```

## Fill whether the visit has mri, pet, dti or not

```python
def str2datetime(input):
    return datetime.strptime(input, "%Y-%m-%d")
```

```python
path = "D:/0.PhD/Project/ADPrediction/src/dataset/ADNI/renamed"
data_preprocess = data_filled_dx
list_ptid = list(data_preprocess["PTID"].unique())
for ptid in list_ptid:
#     if "168_S_6151" != ptid:
#         continue
    print(ptid)
    path_ptid = os.path.join(path, ptid)
    list_visit_mri = [filename for filename in os.listdir(path_ptid) if "MRI" in filename]
    list_visit_pet = [filename for filename in os.listdir(path_ptid) if "PET" in filename]
    list_visit_dti = [filename for filename in os.listdir(path_ptid) if "DTI" in filename]
    patientInfo = np.array(data_preprocess[["EXAMDATE", "Month", "AGE"]][data_preprocess["PTID"] == ptid].sort_values(by="Month", ascending=True))
    rid = data_preprocess[data_preprocess["PTID"] == ptid].index[0]
    #####
    for idx in range(len(patientInfo)):
        examdate = str2datetime(patientInfo[idx, 0])
        month = patientInfo[idx, 1]
    
        if month == 3:
            continue
        elif month == 0:
            baselinedate = examdate
            age_bl = patientInfo[idx, 2]
        else:
            data_preprocess["AGE"].loc[rid][data_preprocess.loc[rid]["Month"] == month] = age_bl + abs(baselinedate - examdate).total_seconds()/60/60/24/365
        # MRI
        
        list_diff_mri = []
        for i, filename in enumerate(list_visit_mri):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3 : 
                list_diff_mri.append(filename)
        if len(list_diff_mri) > 0:
            print("having MRI at: ", month, list_diff_mri[0])
            if len(patientInfo) == 1:
                data_preprocess.at[rid,'MRI'] = 1
            else:
                data_preprocess["MRI"].loc[rid][data_preprocess.loc[rid]["Month"] == month] = 1

        ## PET
        list_diff_pet = []
        for i, filename in enumerate(list_visit_pet):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3 or (diff <= 5 and "168_S_6151" == ptid):
                list_diff_pet.append(filename)
        if len(list_diff_pet) > 0:
            print("having PET at: ", month, list_diff_pet)
            if len(patientInfo) == 1:
                data_preprocess.at[rid,'PET'] = 1
            else:
                data_preprocess["PET"].loc[rid][data_preprocess.loc[rid]["Month"] == month] = 1
        ### DTI
        list_diff_dti = []
        for i, filename in enumerate(list_visit_dti):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3:
                list_diff_dti.append(filename)

        if len(list_diff_dti) > 0:
            print("having DTI at: ", month, list_diff_dti[:2])
            if len(patientInfo) == 1:
                data_preprocess.at[rid,'DTI'] = 1
            else:
                data_preprocess["DTI"].loc[rid][data_preprocess.loc[rid]["Month"] == month] = 1
    print()
```
