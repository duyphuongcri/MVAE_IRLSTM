```python
import os
import shutil
import pandas as pd
import numpy as np 
from datetime import datetime
```

```python
data = pd.read_csv("../dataset/ADNI/data_cleaned.csv")
list_ptid = list(data["PTID"].unique())
print(len(list_ptid))
data
```

```python
def str2datetime(input):
    return datetime.strptime(input, "%Y-%m-%d")
```

```python
path_root = "G:/Dataset/ADNI/3_smooth (SPM)"
path2save = "G:/Dataset/ADNI/TrainingData"


for k, ptid in enumerate(list_ptid):
#     if "168_S_6151" != ptid:
#         continue
    print(k, ptid)

    path_ptid = os.path.join(path_root, ptid)
    if not os.path.exists(os.path.join(path2save, ptid)):
        os.mkdir(os.path.join(path2save, ptid))
    
    list_visit_mri = [filename for filename in os.listdir(path_ptid) if "MRI" in filename]
    list_visit_pet = [filename for filename in os.listdir(path_ptid) if "PET" in filename]
    list_visit_dti = [filename for filename in os.listdir(path_ptid) if "DTI" in filename]
    patientInfo = np.array(data[["EXAMDATE", "Month", "AGE", "MRI", "PET", "DTI"]][data["PTID"] == ptid].sort_values(by="Month", ascending=True))
    rid = data[data["PTID"] == ptid].index[0]
    #####
    for idx in range(len(patientInfo)):
        examdate = str2datetime(patientInfo[idx, 0])
        month = patientInfo[idx, 1]
    
        if month == 3:
            continue
#         print(month)

        # MRI
        list_diff_mri = []
        for i, filename in enumerate(list_visit_mri):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3:
                list_diff_mri.append(filename)
        if len(list_diff_mri) > 0:
            print("having MRI at: ", month, list_diff_mri[0])
            shutil.copy(os.path.join(path_ptid, list_diff_mri[0]), os.path.join(path2save, ptid, "visit_{:0>3}_{}.nii".format(month, list_diff_mri[0][25:-15])))
#             if not os.path.isfile(os.path.join(path2save, ptid+"_visit_{:0>3}_{}".format(month, list_diff_mri[0][24:]))):
#                 shutil.copy(os.path.join(path_ptid, list_diff_mri[0]), os.path.join(path2save, ptid+"_visit_{:0>3}_{}".format(month, list_diff_mri[0][24:])))
        ## PET
        list_diff_pet = []
        for i, filename in enumerate(list_visit_pet):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3 or (diff <= 5 and "168_S_6151" == ptid):
                list_diff_pet.append(filename)
        if len(list_diff_pet) > 0:
            print("having PET at: ", month, list_diff_pet)
            for k in range(len(list_diff_pet)):
                shutil.copy(os.path.join(path_ptid, list_diff_pet[k]), os.path.join(path2save, ptid, "visit_{:0>3}_{}.nii".format(month, list_diff_pet[k][25:-15])))
#                 shutil.copy(os.path.join(path_ptid, list_diff_pet[k]), os.path.join(path2save, ptid+"_visit_{:0>3}_{}".format(month, list_diff_pet[k][24:])))
        ### DTI
        list_diff_dti = []
        for i, filename in enumerate(list_visit_dti):
            date_collected = str2datetime(filename[-14:-4])
            diff = abs(examdate - date_collected).total_seconds()/60/60/24/30
            if diff <= 3:
                list_diff_dti.append(filename)

        if len(list_diff_dti) > 0:
            print("having DTI at: ", month, list_diff_dti[:2])
            for k in range(2):
                shutil.copy(os.path.join(path_ptid, list_diff_dti[k]), os.path.join(path2save, ptid, "visit_{:0>3}_{}.nii".format(month, list_diff_dti[k][25:-15])))
#                 shutil.copy(os.path.join(path_ptid, list_diff_dti[k]), os.path.join(path2save, ptid+"_visit_{:0>3}_{}".format(month, list_diff_dti[k][24:])))
    print()
```
