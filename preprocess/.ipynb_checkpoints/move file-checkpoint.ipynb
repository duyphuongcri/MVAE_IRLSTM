```python
import os
import shutil
import pandas as pd
```

```python
path = "E:/Dataset/ADNI/0_modified_name"
path2save = "E:/Dataset/ADNI"

list_ptid = os.listdir(path)
print(len(list_ptid))
for ptid in list_ptid:
    print(ptid)
    
    path_ptid = os.path.join(path, ptid)
    path2save_cor = os.path.join(path2save +"/1_co_register", ptid)
    path2save_nor =  os.path.join(path2save +"/2_normalized", ptid)
    path2save_smo = os.path.join(path2save +"/3_smooth", ptid)
    ################
    if not os.path.exists(path2save_cor):
        os.mkdir(path2save_cor)
    if not os.path.exists(path2save_nor):
        os.mkdir(path2save_nor)
    if not os.path.exists(path2save_smo):
        os.mkdir(path2save_smo)
    ####################
    files = os.listdir(path_ptid)
    for filename in files:
        if filename.startswith("r"):
            shutil.move(os.path.join(path_ptid, filename), os.path.join(path2save_cor, filename))
        elif filename.startswith("w"):
            shutil.move(os.path.join(path_ptid, filename), os.path.join(path2save_nor, filename))
        elif filename.startswith("sw"):
            shutil.move(os.path.join(path_ptid, filename), os.path.join(path2save_smo, filename))
            
```
